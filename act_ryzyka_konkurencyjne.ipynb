```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from lifelines import CoxPHFitter
```

```python
df = pd.read_excel('Leaders.xlsx')
print(df.head())
print(df.isnull().sum())
```

```python
df = df.dropna()

variables = ['manner', 'start', 'military', 'age', 'conflict',
            'loginc', 'growth', 'pop', 'land', 'literacy']



# Model 1: lost(0) — zdarzenie = 1, 2 lub 3
# m1 = fit_competing_cox(leaders, [1, 2, 3], 'Model 1: lost(0) — wszystkie zdarzenia')

# # Model 2: lost(0,1,2) — zdarzenie = 3 (non-constitutional)
# m2 = fit_competing_cox(leaders, [3], 'Model 2: lost(0,1,2) — zdarzenie: non-constitutional (type 3)')

# # Model 3: lost(0,1,3) — zdarzenie = 2 (nationalized)
# m3 = fit_competing_cox(leaders, [2], 'Model 3: lost(0,1,3) — zdarzenie: nationalized (type 2)')

# # Model 4: lost(0,2,3) — zdarzenie = 1 (consolidated)
# m4 = fit_competing_cox(leaders, [1], 'Model 4: lost(0,2,3) — zdarzenie: consolidated (type 1)')
```

```python
cph = CoxPHFitter()

df['event'] = df['lost'].isin([1,2,3])

cph.fit(
    df[variables + ['years', 'event', 'region']],
    duration_col='years',
    event_col='event',
    strata=['region']
)

cph.print_summary(decimals=3)
```
